```python
from urllib.request import urlopen

url = 'https://alura-site-scraping.herokuapp.com/index.php'

response = urlopen(url)
html = response.read()
html
```

```python
type(html)
```

## Convertendo de bytes para string

```python
html = html.decode('utf-8')
```

```python
type(html)
```

## Eliminando caracteres de tabulação, quebra de linha etc

```python
" ".join(html.split())
```

```python
# Para que o BeautifulSoup reconheça as tags, precisa eliminar espaços entre os demarcadores
" ".join(html.split()).replace('> <', '><')
```

## Organizando

```python
def trata_html(input):
    return " ".join(input.split()).replace('> <', '><')
```

```python
html = trata_html(html)
html
```
