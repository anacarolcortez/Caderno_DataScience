# Teste de Hipótese

### Correlação entre variáveis categóricas

#### Hipótese Nula
<p> A distribuição da taxa da gorjeta é igual nos dois grupos </p>

#### Hipótese Alternativa
<p> A distribuição da taxa da gorjeta não é igual nos dois grupos </p>

```python
from scipy.stats import ranksums
```

```python
sobremesa = tips.query("Sobremesa == 'sim'").Porcentagem
sem_sobremesa = tips.query("Sobremesa == 'não'").Porcentagem
```

```python
r = ranksums(sobremesa, sem_sobremesa)
print(f'O valor do p-value é {r.pvalue}')
```

<p>O pvalue de 0.52 informa que, apesar de os dois grupos terem dado gorjetas de valores diferentes entre si, a diferença é insignificante do ponto de vista matemático. Para aceitar essa hipótese, o pvalue deveria ser menor ou igual a 0.05</p>

<p>Desta forma, descarta-se a  hipótese alternativa, aceitando apenas a nula.</p>

## Teste de Hipótese

#### Hipótese Nula
<p> A distribuição do valor da conta é igual no sábado e no domingo </p>

#### Hipótese Alternativa
<p> A distribuição do valor da conta não é igual no sábado e no domingo </p>

```python
conta_sabado = tips.query("Dia_Semana == 'sábado'").Valor_Conta
conta_domingo = tips.query("Dia_Semana == 'domingo'").Valor_Conta
```

```python
r2 = ranksums(conta_sabado, conta_domingo)
print(f'O valor do p-value é {r2.pvalue}')
```

##### Aceita-se a hipótese nula (pvalue superior a 0.05)
