```python
import nltk
from nltk.tokenize import word_tokenize 
```

```python
with open('dados/artigos.txt', 'r') as f:
    artigos = f.read()
```

```python
print(artigos[:500])
```

## Tokenização

```python
tokens = nltk.tokenize.word_tokenize(artigos)
```

```python
print(tokens[:50])
```

```python
print(f'Quantidade de tokens: {len(tokens)}')
```

```python
def filtra_palavras(lista_tokens):
    lista_palavras = []
    for token in lista_tokens:
        if token.isalpha():
            lista_palavras.append(token.lower())
    return lista_palavras
```

```python
lista_palavras = filtra_palavras(tokens)
print(lista_palavras[:50])
```

```python
print(f'Quantidade de palavras: {len(lista_palavras)}')
```

```python
print(f'Quantidade de palavras únicas: {len(set(lista_palavras))}')
```

## Construindo o algoritmo de correção

```python
exemplo = "lgica"
```

```python
def insere_letras(palavras_fatiadas):
    novas_palavras = []
    letras = 'abcdefghijklmnopqrstuvwxyzáâãéêíóôõúç'
    for E, D in palavras_fatiadas:
        for letra in letras:
            novas_palavras.append(E + letra + D)
    return novas_palavras
```

```python
def gerador_palavras(palavra):
    fatias = []
    for i in range(len(palavra)+1):
        fatias.append((palavra[:i], palavra[i:]))
    palavras_fatiadas = insere_letras(fatias)
    return palavras_fatiadas
```

```python
print(gerador_palavras(exemplo))
```

<p> É importante otimizar o algoritmo, visto que a palavra certa pode aparecer também em uma versão com a grafia errada no corpus (lista total de palavras).</p> 
<p>Ex: lógica x logica. A probabilidade de "lógica" aparecer, contudo, é maior. </p>

```python
frequencia = nltk.FreqDist(lista_palavras)
total_palavras = len(lista_palavras)
frequencia.most_common(10)
```

```python
def probabilidade(palavra):
    return frequencia[palavra]/total_palavras
```

```python
probabilidade('lógica')
```

```python
probabilidade('lagica')
```

```python
probabilidade('logica')
```

```python
def corretor(palavra):
    palavras_geradas = gerador_palavras(palavra)
    palavra_correta = max(palavras_geradas, key=probabilidade)
    return palavra_correta
```

```python
corretor(exemplo)
```

## Testando o modelo

```python
def cria_dados_teste(arquivo):
    lista_palavras_teste = []
    f = open(arquivo, 'r')
    for linha in f:
        correta, errada = linha.split()
        lista_palavras_teste.append((correta, errada))
    f.close()
    return lista_palavras_teste
```

```python
lista_testes = cria_dados_teste('dados/palavras.txt')
```

```python
def teste_corretor(lista_testes):
    acertos = 0
    total_testes = len(lista_testes)
    taxa_acerto = 0.0
    
    for certa,errada in lista_testes:
        palavra_corrigida = corretor(errada)
        if palavra_corrigida == certa:
            acertos += 1
            
    taxa_acerto = round(acertos*100/total_testes, 2)
    print(f'Taxa de acerto: {taxa_acerto}%')
```

```python
teste_corretor(lista_testes)
```

## Otimizando o algoritmo de correção

```python
def deleta_caracter(palavras_fatiadas):
    novas_palavras = []
    for E, D in palavras_fatiadas:
        novas_palavras.append(E + D[1:])
    return novas_palavras
```

```python
def gerador_palavras(palavra):
    fatias = []
    for i in range(len(palavra)+1):
        fatias.append((palavra[:i], palavra[i:]))
    palavras_fatiadas = insere_letras(fatias)
    palavras_fatiadas += deleta_caracter(fatias)
    return palavras_fatiadas
```

# Código final: algoritmo de correção ortográfica

```python
def insere_letras(palavras_fatiadas):
    novas_palavras = []
    letras = 'abcdefghijklmnopqrstuvwxyzáâãéêíóôõúç'
    for E, D in palavras_fatiadas:
        for letra in letras:
            novas_palavras.append(E + letra + D)
    return novas_palavras

def deleta_caracter(palavras_fatiadas):
    novas_palavras = []
    for E, D in palavras_fatiadas:
        novas_palavras.append(E + D[1:])
    return novas_palavras

def troca_letra(palavras_fatiadas):
    novas_palavras = []
    letras = 'abcdefghijklmnopqrstuvwxyzáâãéêíóôõúç'
    for E, D in palavras_fatiadas:
        for letra in letras:
            novas_palavras.append(E + letra + D[1:])
    return novas_palavras

def inverte_letra(palavras_fatiadas):
    novas_palavras = []
    for E, D in palavras_fatiadas:
        if len(D) > 1:
            novas_palavras.append(E + D[1] + D[0] + D[2:])
    return novas_palavras

def gerador_palavras(palavra):
    fatias = []
    for i in range(len(palavra)+1):
        fatias.append((palavra[:i], palavra[i:]))
    palavras_fatiadas = insere_letras(fatias)
    palavras_fatiadas += deleta_caracter(fatias)
    palavras_fatiadas += troca_letra(fatias)
    palavras_fatiadas += inverte_letra(fatias)
    return palavras_fatiadas
```
