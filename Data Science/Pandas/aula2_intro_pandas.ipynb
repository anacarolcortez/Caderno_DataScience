```python
import pandas as pd
```

```python
imoveis = pd.read_csv('aluguel/aluguel.csv', sep=';')
```

```python
imoveis.head()
```

```python
print(f'A base de dados apresenta {imoveis.shape[0]} registros e {imoveis.shape[1]} colunas')
```

```python
print('Tipos de dados')
imoveis.dtypes
```

### Criando um dataframe

```python
tipos_de_imoveis = imoveis['Tipo'].drop_duplicates()
```

```python
tipos_de_imoveis = pd.DataFrame(tipos_de_imoveis)
```

```python
tipos_de_imoveis.index = ["imov" + str(i) for i in range(tipos_de_imoveis.shape[0])]
```

```python
tipos_de_imoveis.index.name = "Id"
tipos_de_imoveis
```

# Imóveis Residenciais

```python
imoveis_residenciais = imoveis.query("Tipo == 'Quitinete' or Tipo == 'Casa' or Tipo == 'Apartamento' or Tipo == 'Casa de Vila' or Tipo == 'Casa de Condomínio'")
```

```python
imoveis_residenciais.index = range(imoveis_residenciais.shape[0])
```

```python
imoveis_residenciais.index.name = "Id"
```

```python
imoveis_residenciais
```

### Exportando o resultado para arquivo csv

```python
imoveis_residenciais.to_csv('aluguel/aluguel_residencial.csv', sep=';', index=False)
```
