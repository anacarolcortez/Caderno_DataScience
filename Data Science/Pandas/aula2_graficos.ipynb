```python
import pandas as pd
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.rc('figure', figsize = (15,8))
```

```python
imoveis = pd.read_csv('aluguel/aluguel_residencial_sem_outliers.csv', sep=';')
```

```python
area = plt.figure() 
```

```python
#Subdividindo a área do gráfico em 4 partes:
g1 = area.add_subplot(2, 2, 1)
g2 = area.add_subplot(2, 2, 2)
g3 = area.add_subplot(2, 2, 3)
g4 = area.add_subplot(2, 2, 4)
```

```python
g1.scatter(imoveis.Valor, imoveis.Area)
g1.set_title('Valor X Área')

g2.hist(imoveis.Valor)
g2.set_title('Histograma')

dados_g3 = imoveis.Valor.sample(100) 
dados_g3.index = range(dados_g3.shape[0])
g3.plot(dados_g3)
g3.set_title('Amostra (Valor)')

grupo = imoveis.groupby('Tipo')['Valor']
label = grupo.mean().index
valores = grupo.mean().values
g4.bar(label, valores)
g4.set_title('Valor Médio por Tipo')
```
