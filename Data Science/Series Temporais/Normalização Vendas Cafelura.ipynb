# Vendas Cafelura

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
cafelura = pd.read_csv('dados/cafelura.csv')
cafelura.head()
```

```python
cafelura.dtypes
```

```python
cafelura['mes'] = pd.to_datetime(cafelura['mes'])
cafelura.dtypes
```

```python
print('Quantidade de linhas e colunas:', cafelura.shape)
print('Quantidade de dados nulos:', cafelura.isna().sum().sum())
```

```python
def grafico_linha(data, x, y, title, xlabel, ylabel):
    ax = sns.lineplot(x=x, y=y, data=data)
    ax.figure.set_size_inches(12,6)
    ax.set_title(title, loc='left', fontsize=18)
    ax.set_xlabel(xlabel, fontsize=14)
    ax.set_ylabel(ylabel, fontsize=14)
    ax = ax
```

```python
grafico_linha(cafelura, 'mes', 'vendas', 'Vendas da Cafelura de 2017 e 2018', 
              'Tempo', 'Vendas')
```

```python
quantidade_de_dias_de_fds = pd.read_csv('dados/dias_final_de_semana.csv')
quantidade_de_dias_de_fds.head()
```

```python
quantidade_de_dias_de_fds['quantidade_de_dias'].values
```

```python
cafelura['vendas_normalizadas'] = cafelura['vendas'] / quantidade_de_dias_de_fds['quantidade_de_dias'].values
cafelura.head()
```

```python
grafico_linha(cafelura, 'mes', 'vendas_normalizadas',
       'Vendas normalizadas da Cafelura de 2017 a 2018',
      'Tempo', 'Vendas normalizadas')
```

```python
plt.figure(figsize=(12,8))
ax = plt.subplot(2,1,1)
ax.set_title('Vendas Cafelura 2017 e 2018', fontsize=18)
sns.lineplot(x='mes', y='vendas', data=cafelura)
ax = plt.subplot(2,1,2)
ax.set_title('Vendas Normalizadas Cafelura 2017 e 2018', fontsize=18)
sns.lineplot(x='mes', y='vendas_normalizadas', data=cafelura)
ax = ax
```
